1.What are the different products available for the customer to use

## Data:
The data contains Model of the treadmill,customer charateristics,Usage,miles run and self rated fitness score of the customers

1.Product - the model no. of the treadmill

2.Age - in no of years, of the customer

3.Gender - of the customer

4.Education - in no. of years, of the customer

5.Marital Status - of the customer

6.Usage - Avg. # times the customer wants to use the treadmill every week

7.Fitness - Self rated fitness score of the customer (5 - very fit, 1 - very unfit)

8.Income - of the customer

9.Miles- expected to run

### Data Preprocessing

```python
#Categorical variables Product,Marital Status and Gender need to be converted to categorical type

cardio['Product']=cardio['Product'].astype('category')

cardio['MaritalStatus']=cardio['MaritalStatus'].astype('category')

cardio['Gender']=cardio['Gender'].astype('category')
```

```python
#Converting Education to categorical type which will help to analyse it better

cardio['Education']=cardio['Education'].astype('category')
```

### Analysing the age of customers 

```python
sns.distplot(cardio['Age'],color='g');
mean= plt.axvline(np.mean(cardio['Age']),color='r', linestyle='--')#plotting a mean line through graph
median= plt.axvline(np.median(cardio['Age']),color='b', linestyle='-')#plotting a median line through graph

plt.legend({'Mean':mean,'Median':median})# plotting legend for mean and median
print('The Mean Age of customers is',int(cardio['Age'].mean()))
print('The median Age of customers is ',int(cardio['Age'].median()))

```

## Multivariate Analysis

### Checking the correlation of the numerical variables using a heatmap

```python
plt.figure(figsize=(10,5))# Setting the fig size of the plot
sns.heatmap(cardio.corr(), annot=True, linewidths=.5, fmt= '.1f', center = 1 ) 
plt.show()# showing the plot
```

Observations:
    
1.Fitness and Miles run per week have high correlation

2.Usage and Miles run per week have high correlation

```python
#Analysing the Gender and usage per week
sns.swarmplot(cardio['Gender'],cardio['Usage']);
```

### Analysis of the usage of the Products

```python
#Grouping the products using groupby function

cardio.groupby(by=['Product']).sum().reset_index().sort_values(['Usage']).plot(x='Product', y='Usage',kind='bar',
                                                                                                           figsize=(5,5))
plt.title('Product vs Usage')#title of the graph
plt.show()
```

Observation: The product TM195 is the most used product among the three products

### Analysis of the product vs the self rated fitness score of the customer

```python
#Grouping the products using groupby function
cardio.groupby(by=['Product']).sum().reset_index().sort_values(['Fitness']).plot(x='Product', y='Fitness',kind='bar',
                                                                                                           figsize=(5,5),color='orange')
plt.title('Product vs Fitness')
plt.show()
```

Observation:
1.The customers who use TM195 product have rated themselves higher for the fitness score
